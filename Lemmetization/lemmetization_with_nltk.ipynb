```python
#Import wordnet
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
```

```python
lemmetizer = WordNetLemmatizer()
```

```python
lemmetizer.lemmatize('studying')
```

```python
lemmetizer.lemmatize('studying', pos=wordnet.VERB)
```
