```python
# Perform stanard imports
import spacy
nlp=spacy.load("en_core_web_sm")
```

```python
# From spacy Basics:
doc=nlp(u'This is the first sentence. This is another sentence. This is the last sentence.')

for sent in doc.sents: #sents is a generator that yields sentences
    print(sent)
```

```python
# parsing the segementation start token happens during nlp pipeline
doc2=nlp(u'This is a sentence. This is a sentence. This is a sentence.')
for token in doc2:
    print(token.is_sent_start, ''+token.text)
```

```python
# spaCy's default behavior 
doc3=nlp(u'"Management is doing this right; leadership is doing the right things."- Peter Drucker.')
for sent in doc3.sents:
    print(sent)
```

```python
#adding a new rule to the pipeline
from spacy.language import Language
@Language.component("set_custom_boundaries")
def set_custom_boundaries(doc):
  for token in doc[:-1]:
    if token.text == ';':
          doc[token.i+1].is_sent_start = True
  return doc
nlp.add_pipe("set_custom_boundaries",before='parser')
nlp.pipe_names
```

```python
# Re-run the doc object creation:
doc4=nlp(u'"Management is doing this right; leadership is doing the right things."- Peter Drucker.')
for sent in doc4.sents:
    print(sent)
```
