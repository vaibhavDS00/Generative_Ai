```python
name='Freed'
#Using the old .format() method:
print ('His name is {var}.'.format(var=name))
#using f-strings:
print(f'His name is {name}.')
```

```python
#To Highlight the s-string variable
print(f'His name is {name!r}')
```

```python
d= {'a':123,'b':456}
print(f"My address: {d['a']} main street")
```

```python

library=[('Author','Topic','Pages'),('Twain','Rafting',601),('Feynman','Physics',95),('Halmilton','Mythology',114)]
for book in library:
    print(f'{book[0]:{10}} {book[1]:{10}} {book[2]:{5}}')
```

```python

library=[('Author','Topic','Pages'),('Twain','Rafting',601),('Feynman','Physics',95),('Halmilton','Mythology',114)]
for book in library:
    print(f'{book[0]:{10}} {book[1]:{10}} {book[2]:*>{7}}')
```

```python
from datetime import datetime
today=datetime(year=2018,month=1,day=27)
print(f'{today:%B,%d,%Y}')
```

```python
from datetime import datetime
today=datetime(year=2018,month=1,day=27)
print(f'{today:%b,%d,%y}')
```

```python
from datetime import datetime
today=datetime(year=2018,month=1,day=27)
print(f'{today:%D}')
```
