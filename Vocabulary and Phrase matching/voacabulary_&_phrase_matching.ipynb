```python
# Perform standard imports 
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
#import the Matcher Library
from spacy.matcher import Matcher
matcher=Matcher(nlp.vocab)
```

```python
# Creating patterns
pattern1 = [{'LOWER': 'solarpower'}]
pattern2 =[{'LOWER': 'solar'},{'LOWER': 'power'}]
pattern3 =[{'LOWER': 'solar'},{'IS_PUNCT': True},{'LOWER': 'power'}]
```

```python
matcher.add('SolarPower',[pattern1, pattern2, pattern3])
```

```python
doc=nlp(u'The Solar Power imdustry continues to grow as demand for solarpower increases. Solar-Power cars are gaining popularity.')
```

```python
found_matches=matcher(doc)
print(found_matches)
```

```python
for match_id, start, end in found_matches:
    string_id = nlp.vocab.strings[match_id]
    span = doc[start:end]
    print(match_id, string_id, start, end, span.text)
```

```python
# redefine the patterns:
pattern1= [{'LOWER': 'solarpower'}]
pattern2= [{'LOWER': 'soalr'}, {'IS_PUNCT': True, 'OP':'*'},{'LOWER': 'power'}]
```

```python
# Add the new set of patterns to the 'SolarPower' matcher:
matcher.add('SolarPower',[pattern1,pattern2])
```

```python
found_matches=matcher(doc)
print(found_matches)
```

```python
# Be caefull with lemmas
pattern1= [{'LOWER': 'solarpower'}]
pattern2= [{'LOWER': 'solar'}, {'IS_PUNCT': True, 'OP':'*'},{'LEMMA': 'power'}]
```

```python
# Remove the old patterns to avoid duplication:
matcher.remove('SolarPower')
```

```python
matcher.add('SolarPower', [pattern1,pattern2])
```

```python
doc2=nlp(u'Solar-Powered energy runs solar-powered cars.')
```

```python
found_matches=matcher(doc2)
print(found_matches)
```
