```python
# Often when searching text for a certain keyword, it helps if the search returns variations of the word. For instance,
# searching for "boat" might also return "boats" and "boating". Here, 
# "boat" would be the stem for [boat, boater, boating,Â boats].
```

```python
# Import the toolkit and full Porter Stemmer library
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
words = ["run", "runner", "running", "ran", "easily", "fairly"]
```

```python
for word in words:
    print(word+'-->'+ p_stemmer.stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer

# The snowball Stemmer requires that you pass a langugage parameter
s_stemmer = SnowballStemmer(language='english')
```

```python
words = ['run', 'runner', 'running', 'ran', 'easily', 'fairly']
for word in words:
    print(word+'-->'+ s_stemmer.stem(word))
```

```python
words = ['generous', 'generation', 'generously', 'generate']
for word in words:
    print(word+'-->'+ s_stemmer.stem(word))
```

```python
word=['consolingly']
for w in word:
    print(w+'-->'+ s_stemmer.stem(w))

for w in word:
    print(w+'-->'+ p_stemmer.stem(w))
```

```python
pharse='I am meeting him tommorrow at the meeting'
for w in pharse.split():
    print(w+'-->'+ p_stemmer.stem(w))
print('---------------------')
for w in pharse.split():
    print(w+'-->'+ s_stemmer.stem(w))
```
