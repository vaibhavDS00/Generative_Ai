```python
s="I like this course."
```

```python
s.split() # It uses white space to split the sentence into words.
```

```python
# Character based tokenization
```

```python
list = []
for char in s:
    list.append(char)
print(list)
```

```python
# Remove all white spaces from the string
text="".join(s.split())
```

```python
text
```

```python
list2 = []
for char in text:
    list2.append(char)
print(list2)
```

```python
# Subword based tokenization
```

```python
string = "I am learning Tokenization"
```

```python
newString = string.split()
```

```python
newString
```

```python
list3 = []
for word in newString:
    if word[-3:] == "ing":
        list3.append(word.partition("ing"))
    elif word[-7:] == "ization":
        list3.append(word.partition("ization"))
    else:
        list3.append(word)
print(list3)
```

```python
list4=[]
for element in list3:
    if type(element) is tuple:
        list4.append(element[0])
        list4.append(element[1])
    else:
        list4.append(element)
print(list4)
```
