```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
# NETK isan open source library. It provides many functionalities and include pre-trained model
# NETK integrates well with other libraries, such as seikit-learn and Ten-sorFlow, which can be useful if you need to build more advanced models.
# For research or educational purposes, NETK is a popular choice among researchers and edueators due to its flexibility, extensive documentation.
```

```python
# Pre-trainded model for tokenization
nltk.download('punkt_tab')
```

```python
sentence = "Today is friday. Next week, we're coming from U.S.!"
```

```python
for token in word_tokenize(sentence):
    print(token)
```

```python
sentence.split()
```
